```python
birth2008 = pd.read_csv('../birth/birthdata/birth_2008_1.csv')
birth2009 = pd.read_csv('../birth/birthdata/birth_2009_1.csv')
birth2010 = pd.read_csv('../birth/birthdata/birth_2010_1.csv')
birth2011 = pd.read_csv('../birth/birthdata/birth_2011_1.csv')
birth2012 = pd.read_csv('../birth/birthdata/birth_2012_1.csv')
birth2013 = pd.read_csv('../birth/birthdata/birth_2013_1.csv')
birth2014 = pd.read_csv('../birth/birthdata/birth_2014_1.csv')
birth2015 = pd.read_csv('../birth/birthdata/birth_2015_1.csv')
birth2016 = pd.read_csv('../birth/birthdata/birth_2016_1.csv')
birth2017 = pd.read_csv('../birth/birthdata/birth_2017_1.csv')
```

```python
birth = pd.merge(birth2008,table, how='outer')
birth = pd.merge(birth2009, birth, how='outer')
birth = pd.merge(birth2010, birth, how='outer')
birth = pd.merge(birth2011, birth, how='outer')
birth = pd.merge(birth2012, birth, how='outer')
birth = pd.merge(birth2013, birth, how='outer')
birth = pd.merge(birth2014, birth, how='outer')
birth = pd.merge(birth2015, birth, how='outer')
birth = pd.merge(birth2016, birth, how='outer')
birth = pd.merge(birth2017, birth, how='outer')
```
