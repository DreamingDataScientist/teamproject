```python
death2008 = pd.read_csv('../death/death_raw/dead_2008_1.csv')
death2009 = pd.read_csv('../death/death_raw/dead_2009_1.csv')
death2010 = pd.read_csv('../death/death_raw/dead_2010_1.csv')
death2011 = pd.read_csv('../death/death_raw/dead_2011_1.csv')
death2012 = pd.read_csv('../death/death_raw/dead_2012_1.csv')
death2013 = pd.read_csv('../death/death_raw/dead_2013_1.csv')
death2014 = pd.read_csv('../death/death_raw/dead_2014_1.csv')
death2015 = pd.read_csv('../death/death_raw/dead_2015_1.csv')
death2016 = pd.read_csv('../death/death_raw/dead_2016_1.csv')
death2017 = pd.read_csv('../death/death_raw/dead_2017_1.csv')

```

```python
table = pd.read_csv('../pro_data/pop_table1.csv')
```

```python
death = pd.merge(death2008,table, how='outer')
death = pd.merge(death2009,death, how='outer')
death = pd.merge(death2010,death, how='outer')
death = pd.merge(death2011,death, how='outer')
death = pd.merge(death2012,death, how='outer')
death = pd.merge(death2013,death, how='outer')
death = pd.merge(death2014,death, how='outer')
death = pd.merge(death2015,death, how='outer')
death = pd.merge(death2016,death, how='outer')
death = pd.merge(death2017,death, how='outer')
```
