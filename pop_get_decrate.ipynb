```python
pr_0801a = p_0801[p_0801.columns[get_idx2]]
pr_0801a['노인'] = pr_0801a[pr_0801a.columns[1:9]].sum(axis=1)
pr_0801a['2030여성'] = pr_0801a[pr_0801a.columns[9:13]].sum(axis=1)
pr_0801a['0802소멸비율'] = pr_0801a['2030여성'] / (pr_0801a['노인'] / 2)
pr_0801a = pr_0801a[pr_0801a.columns[[0, -1]]]
pr_0801a.head()
```

```python
y_f = '{0:02d}'
m_f = '{0:02d}'

for i in range(8,18):
    for j in range(1,11,3):
        p = pd.read_csv('C:/Users/DJA/Downloads/데이터/원본자료/새 폴더/pop_20'+
                        y_f.format(i)+m_f.format(j)+'_20'+y_f.format(i)+m_f.format(j+2)+'_u8.csv')
        p0 = p[p.columns[get_idx]]
        p1 = p[p.columns[get_idx2]]
        p2 = p[p.columns[get_idx3]]
        p0['노인'] = p0[p0.columns[1:9]].sum(axis=1)
        p1['노인'] = p1[p1.columns[1:9]].sum(axis=1)
        p2['노인'] = p2[p2.columns[1:9]].sum(axis=1)
        p0['2030여성'] = p0[p0.columns[9:13]].sum(axis=1)
        p1['2030여성'] = p1[p1.columns[9:13]].sum(axis=1)
        p2['2030여성'] = p2[p2.columns[9:13]].sum(axis=1)
        p0[y_f.format(i)+m_f.format(j)+'소멸비율'] = p0['2030여성'] / (p0['노인'] / 2)
        p1[y_f.format(i)+m_f.format(j+1)+'소멸비율'] = p1['2030여성'] / (p1['노인'] / 2)
        p2[y_f.format(i)+m_f.format(j+2)+'소멸비율'] = p2['2030여성'] / (p2['노인'] / 2)
        p0 = p0[p0.columns[[0, -1]]]
        p1 = p1[p1.columns[[0, -1]]]
        p2 = p2[p2.columns[[0, -1]]]
        p0 = pd.merge(p0, p1, how='outer')
        p0 = pd.merge(p0, p2, how='outer')
        pr_0801 = pd.merge(pr_0801, p0, how='outer')
pr_0801.head()
```
