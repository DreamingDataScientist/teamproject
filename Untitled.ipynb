## 인구소멸지역을 활용한 폐교현황예측

```r
school <- read.csv('C:\\Users\\DJA\\Desktop\\data\\2013_폐교현황.csv', header = T,stringsAsFactor=F)
pop <- read.csv('C:\\Users\\DJA\\Desktop\\data\\pop2013.csv', header=T,  stringsAsFactor=F)
```

```r
school_r <- paste(school$도,school$군.시,school$읍.면.구,sep=" ")
pop_r <- paste(pop$도,pop$시.군.구,pop$읍면동,sep=" ")
```

```r
sch_2013 <- data.frame('지역'=school_r,'폐교수'=school$폐교현황)
pop_2013 <- data.frame('지역'=pop_r, '소멸비율'=pop$소멸비율)
head(pop_2013)
```

```r
library(plyr)
result <- join(sch_2013, pop_2013, type='right')
head(result)
```
