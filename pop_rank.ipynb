```python
import pandas as pd
pr_08 = pd.read_csv('C:/Users/DJA/Downloads/데이터/정제자료/인구관련/pop_monthly_rate08_17_u8.csv')

pr_l20 = []

form = '{0:02d}{1:02d}'
for i in range(8,18):
     for j in range(1,13):
        df = pr_08[['지역', form.format(i,j)]]
        pr = df.sort_values(by=[form.format(i,j)],axis=0,na_position='last')
        pr_l20 = pr_l20 + pr['지역'].head(20).tolist()
        
pr_l20

```

```python
count = {}
for i in pr_l20:
    try: count[i] += 1
    except: count[i]=1

count
```
