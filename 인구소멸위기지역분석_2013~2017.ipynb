## Index를 지역 시군구로 정함

```python
pop2013_1.index = pop2013[['시군구']]
pop2013_1.head()
```

## 노인 인구 65세이상의 절반보다 여성20-30대가 더 적을 경우 소멸위기지역이며 소멸비율을 계산 
- 소멸비율이 1보다 작을 경우 위험한 지역, 1이상일 경우 소멸 위험 지역에 해당이 안됨

```python
pop2013_1.is_copy = False

pop2013_1['소멸비율'] = pop2013_1['여성_20-39세'] / (pop2013_1['총_65세이상'] / 2)
pop2013_1.head()
```

## 소멸비율이 1보다 작은 지역일 경우 소멸위기지역이라고 새로운 열을 생성 

```python
pop2013_1['소멸위기지역'] = pop2013_1['소멸비율'] < 1.0
pop2013_1.head()
```

## 2013년 가장 소멸 비율이 높은 지역 30위까지 보여줌

```python
data2013 = pop2013_1.sort_values(by='소멸비율', ascending=True)
data2013.head(30)
```
